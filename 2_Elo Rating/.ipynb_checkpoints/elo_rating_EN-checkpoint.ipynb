# Predicting NBH winner using *Elo rating*
[Elo rating system](https://en.wikipedia.org/wiki/Elo_rating_system) is a method to calculate the relative skill levels developed by Arpad Elo. Long story short, we can use this method to predict who will win the next NBH competition based on the competing chef's historical win-loss record. </br>

I referenced code from [Elo Ratings in Python from Kaggle](https://www.kaggle.com/kplauritzen/elo-ratings-in-python/notebook) using my own compiled dataset on NBH.

## What is Elo Rating? (a short version)

This is my best attempt to explain Elo rating in three sentences or less. 
1. Every player starts with a score card (with number 100 inscribed) hanging around their neck.
2. At every match, the winner takes few score from the loser.
3. If it is an unexpected win (say *David's victory over Goliath*), the winner takes **more** score than as it were the other way around.

![elo_pokemon](elo_pokemon.png)

Simple enough? Now we are down with what Elo rating is, let's use it on our NBH chefs.

## 1. Load game dataset

```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
from matplotlib import font_manager, rc
from collections import Counter
%matplotlib inline
```

Load pickle file with data on previous match data and the dates when the episode was aired. 

```python
with open("game_data.pkl", 'rb') as pk:
    game_data = pickle.load(pk)
```

```python
df = game_data[["win", "lose", "date"]]
df[:3]
```

## 2. Set constants we neeed to compute Elo ratings

```python
mean_elo = 1500
elo_width = 400
k_factor = 64
```

```python
chef_list = df["win"].tolist() + df["lose"].tolist()
frequency = Counter(chef_list).most_common()
frequency_dict = dict(frequency)
```

```python
# Set chef id dictionary
chef_df = pd.DataFrame({"chef": pd.Series([c[0] for c in frequency]).values, "id": np.arange(len(frequency_dict))})
chef_id_dict = dict(zip(chef_df.chef, chef_df.id))
id_chef_dict = dict(zip(chef_df.id, chef_df.chef))
```

We will embed chef names with IDs so it will be easier to handle.

```python
# Change
df["win"] = df["win"].apply(lambda x: chef_id_dict.get(x, -1))
df["lose"] = df["lose"].apply(lambda x: chef_id_dict.get(x, -1))
```

## 3. Define functions

```python
def expected_result(elo_a, elo_b):
    expect_a = 1.0/(1+10**((elo_b - elo_a)/elo_width))
    return expect_a

def update_elo(winner_elo, loser_elo):
    expected_win = expected_result(winner_elo, loser_elo)
    change_in_elo = k_factor * (1-expected_win)
    winner_elo += change_in_elo
    loser_elo -= change_in_elo
    return winner_elo, loser_elo

def update_end_of_season(elos):
    diff_from_mean = elos - mean_elo
    elos -= diff_from_mean/3
    return elos
```

```python
# Set df columns
df["w_elo_before_game"] =0
df["w_elo_after_game"] =0
df["l_elo_before_game"] =0
df["l_elo_after_game"] =0

# Set time
df["year"] = pd.DatetimeIndex(df['date']).year
df["month"] = pd.DatetimeIndex(df["date"]).month
df["day"] = pd.DatetimeIndex(df["date"]).day
df['total_days'] = (df.year - 2015)*365.25 + (df.month-1)*30 + df.day
df[:3]
```

## 4. Loop and update

```python
n_teams = len(chef_id_dict)
current_elos = np.ones(shape=(n_teams)) * mean_elo
```

```python
# matrix to update chef's elo score
df_chef_elos = pd.DataFrame(index = df.total_days.unique(), columns = range(n_teams))
```

```python
df_chef_elos.iloc[0,:] = current_elos
```

```python
for row in df.itertuples():
        
    idx = row.Index
    w_id = row.win
    l_id = row.lose
    
    # Get current elos
    w_elo_before = current_elos[w_id]
    l_elo_before = current_elos[l_id]
    w_elo_after, l_elo_after = update_elo(w_elo_before, l_elo_before)
    
    # Save updated elos
    df.at[idx, 'w_elo_before_game'] = w_elo_before
    df.at[idx, 'l_elo_before_game'] = l_elo_before
    df.at[idx, 'w_elo_after_game'] = w_elo_after
    df.at[idx, 'l_elo_after_game'] = l_elo_after
    current_elos[w_id] = w_elo_after
    current_elos[l_id] = l_elo_after
    
    # Save elos to team DataFrame
    today = row.total_days
    df_chef_elos.at[today, w_id] = w_elo_after
    df_chef_elos.at[today, l_id] = l_elo_after
```

After the match, chef no. 4 won and we can see that no. 4's score increased by 32, and no.13' score decreased by 32.

## 5. Evaluation
Let's evaluate the result using the recent samples.

```python
n_samples = 50
samples = df[-n_samples:]
loss = 0
expected_list= []
```

```python
for row in samples.itertuples():
    w_elo = row.w_elo_before_game
    l_elo = row.l_elo_before_game
    w_expected = expected_result(w_elo, l_elo)
    expected_list.append(w_expected)
    loss += np.log(w_expected)
print(loss/n_samples)
```

```python
sns.distplot(expected_list, kde=False, bins=20)
plt.xlabel('Elo Expected Wins for Actual Winner')
plt.ylabel('Counts')
plt.title("Expected Wins for Actual Winners")
```

It looks like predictability is not so good (yet).
