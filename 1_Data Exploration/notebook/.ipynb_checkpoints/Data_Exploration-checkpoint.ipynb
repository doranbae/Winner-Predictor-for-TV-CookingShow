```python
import pandas as pd
import copy 
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
file_path = '../data/game_data.csv'
game_data = pd.read_csv(file_path, encoding='utf-8')
```

```python
game_data = game_data.rename(columns = {"Unnamed: 0": "epi_num"})
game_data[:10]
```

Let's create how many times they have competed, and their winning rate.

```python
# Create frequency table and dictionary
chefs = game_data["win"].tolist() + game_data["lose"].tolist()
frequency = Counter(chefs).most_common()
frequency_dict = dict(frequency)
```

```python
chef_df = pd.DataFrame({"chef": pd.Series([c[0] for c in frequency]).values, "id": np.arange(len(frequency_dict))})
```

```python
chef_id_dict = dict(zip(chef_df.chef, chef_df.id))
reverse_chef_id_dict = dict(zip(chef_df.id, chef_df.chef))
```

```python
# Create win_count, loss_count
win_count_dict = dict(Counter(game_data["win"].tolist()).most_common())
lose_count_dict = dict(Counter(game_data["lose"].tolist()).most_common())
```

```python
win_rate = dict()
for chef in chef_df["chef"].tolist():
    win = win_count_dict.get(chef,0)
    num_games = frequency_dict.get(chef,0)
    win_rate[chef] = win/num_games
```

```python
x_label = chef_df["chef"].tolist()
num_games = [frequency_dict.get(x,0) for x in x_label]
win_rate = [win_rate.get(x,0) for x in x_label]
```

```python
def two_scales(ax1, x_axis, data1, data2, c1, c2):
    ax2 = ax1.twinx()
    
    ax1.plot(x_axis, data1, color = c1)
    ax1.set_xlabel("Chefs")
    ax1.set_ylabel('Number of appearances')
    
    ax2.plot(x_axis, data2, color=c2)
    ax2.set_ylabel('Winning rate')
    return ax1, ax2
```

## Who is my nemesis?

```python
game_data_embed = copy.deepcopy(game_data)
# Embed the gamedata with chef ID
game_data_embed["win"] = game_data_embed["win"].apply(lambda x: chef_id_dict.get(x))
game_data_embed["lose"] = game_data_embed["lose"].apply(lambda x: chef_id_dict.get(x))
```

```python
game_data_embed[:3]
```

```python
match_pair = game_data_embed[["win","lose"]].values.tolist()
score_matrix_ = np.zeros([len(x_label),len(x_label)])
#score_matrix_ = np.zeros([15,15])
```

```python
def update_score(matrix, pair):
    try:
        for pair in pair:
            matrix[pair[0], pair[1]] += 1
    except:
        pass
    return matrix
```

```python
score_matrix = update_score(score_matrix_,match_pair)
```

```python
fig, ax = plt.subplots(figsize = (8,8))
cax = plt.imshow(score_matrix[:15,:15], cmap='afmhot')
cbar = fig.colorbar(cax, ticks=[score_matrix.max(),score_matrix.min()], orientation = 'horizontal',fraction=0.046, pad=0.04)
plt.show()
```

```python
#score_matrix_
```

```python
sorted_match_pair = []
for pair in match_pair:
    sort = tuple(sorted(pair))
    sorted_match_pair.append(sort)
```

```python
sorted_match_pair_count = Counter(sorted_match_pair)
sorted_match_pair_dict = dict(sorted_match_pair_count)
```

```python
total_matrix = np.zeros([len(x_label),len(x_label)])
```

```python
for pair in sorted_match_pair_dict.keys():
    p1 = pair[0]
    p2 = pair[1]
    val = sorted_match_pair_dict.get(pair)
    total_matrix[p1,p2] += val
    total_matrix[p2,p1] += val
```

```python
battle_score = np.divide(score_matrix, total_matrix, out=np.zeros_like(score_matrix), where=total_matrix!=0)
```

```python
chef_name_en = ['Kim, Poong','Kim, Sam','Lee, Yeon-bok',"Mihal Ashminov","Choi, Hyun-suk","Jung, Ho-young","Oh, Se-deuk","Lee, Won-il", "Hong, Seok-chun","Lee, Chan-o","Yoo, Hyun-soo","Kim, Raymond",
                "Lee, Jae-hoon","Park, Joon-woo","Jung, Chang-wook"]
```

```python
fig, ax = plt.subplots(figsize = (15,15))

ax = sns.heatmap(battle_score[:15,:15], 
                 annot=True, 
                 fmt=".1f", 
                 linewidths=1.0, 
                 square=1,
                 xticklabels=chef_name_en,
                 yticklabels=chef_name_en,
                 cmap="YlGnBu",
                cbar_kws = {"orientation":"horizontal","shrink": .70})
plt.title('Who is my Nemesis?',fontsize=20)

fig.savefig("../data/nbh_nemesis.png")
```
